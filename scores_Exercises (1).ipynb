### Step 1. Import the necessary libraries

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Step 2. Create the DataFrame that should look like the one below.

```python
raw_data = {'first_name':['Jason','Molly','Tina','Jake','Amy'], 'last_name':['Miller','Jacobson','Ali','Milner','Cooze'], 
            'age':[42,52,36,24,73],'female':[0,1,1,0,1],'preTestScore':[4,24,31,2,3], 'postTestScore':[25,94,57,62,70]}
a = pd.DataFrame(raw_data, columns=['first_name','last_name','age','female','preTestScore','postTestScore'])
a
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(x=a['preTestScore'], y=a['postTestScore'], s=a['age'], alpha=0.5)
axes1.set_title('Pre-Test Score vs Post-Test Score Sized by Age')
axes1.set_xlabel('Pre-Test Score')
axes1.set_ylabel('Post-Test Score')
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(x=a['preTestScore'], y=a['postTestScore'], c=a['female'], s=a['postTestScore']*4.5, alpha=0.5)
axes1.set_title('Pre-Test Score vs Post-Test Score Sized by Post-Test Score and Colored by Sex')
axes1.set_xlabel('Pre-Test Score')
axes1.set_ylabel('Post-Test Score')
```
